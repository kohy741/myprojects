```python
#before splitting data, let's do encoding
from sklearn import preprocessing

features = ['gender', 'work_type', 'ever_married', 'Residence_type', 'smoking_status']
for feature in features:
    le = preprocessing.LabelEncoder()
    le = le.fit(df[feature])
    df[feature] = le.transform(df[feature])
df.head()
```

```python
#let's do Normalization
from sklearn.preprocessing import MinMaxScaler
df_sc = df
scaler = MinMaxScaler()

scaler.fit(df_sc[['avg_glucose_level']])
df_sc['avg_glucose_level'] = scaler.transform(df_sc[['avg_glucose_level']])

scaler.fit(df_sc[['age']])
df_sc['age'] = scaler.transform(df_sc[['age']])

scaler.fit(df_sc[['gender']])
df_sc['gender'] = scaler.transform(df_sc[['gender']])

scaler.fit(df_sc[['work_type']])
df_sc['work_type'] = scaler.transform(df_sc[['work_type']])

scaler.fit(df_sc[['bmi']])
df_sc['bmi'] = scaler.transform(df_sc[['bmi']])

scaler.fit(df_sc[['smoking_status']])
df_sc['smoking_status'] = scaler.transform(df_sc[['smoking_status']])

df_sc.head()
```

```python
#let's find out feature's importance by using decision tree
X=df_sc.drop('stroke', axis = 1)
y = df_sc['stroke']
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
feature_importance = clf.tree_.compute_feature_importances(normalize = True)
print("feature importance = " + str(feature_importance))
```

```python
#let's make a plot to make it easy to see
tree_importances = pd.Series(feature_importance, index=df.drop('stroke',axis=1).columns)
fig, ax = plt.subplots()
tree_importances.plot.bar(feature_importance,subplots=True,sharex=True)
ax.set_title("Feature importances")
for p in ax.patches:
    ax.annotate(str(round(p.get_height(),3)), (p.get_x() * 1.005, p.get_height() * 1.005))
#ax.bar_label(ax.containers[0])
#ax.set_ylabel("Mean decrease in impurity")
fig.tight_layout()
```

```python
# now, it's time to split data.
#first, the data have to split into input data(X) & output data(Y)
X = df_sc.drop(['stroke'], axis = 1) #  dropping the 'stroke' column 
Y = df_sc['stroke'] #   output data(Y) = the 'stroke' column 
```

```python
X.head()
```

```python
Y.head()
```

```python
#and now, split them into test data & training data   ratio 2:8
#split 'X' into  x_train & x_test
#split 'Y' into  y_train & y_test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state=101)

#let's see the shape of them (rows, colunms)
x_train.shape,x_test.shape,  y_train.shape, y_test.shape
```

```python
#let's train with SVM with all features
import time #to measure the training time
from sklearn.svm import SVC
classifier = SVC(random_state = 0, kernel = 'linear')

start = time.time()
classifier.fit(x_train, y_train) #training
stop = time.time()

print(f"Training time: {round(stop - start,3)}s")
```

```python
prediction = classifier.predict(x_test) #predict with the test input
```

```python
#let's compare prediction with y_test to see the result
from sklearn.metrics import plot_confusion_matrix,accuracy_score,precision_score,recall_score,classification_report
import warnings
warnings.filterwarnings('ignore')
print("Accuracy :", accuracy_score(y_test, prediction))
print("Precision :",precision_score(y_test, prediction)) 
print("Recall :",recall_score(y_test, prediction),"\n") 
print(classification_report(y_test, prediction, target_names = ['no_stroke', 'yes_stroke']))
plot_confusion_matrix(classifier, x_test, y_test, cmap='summer')
plt.title('Confusion Matrix')
plt.show()
```
